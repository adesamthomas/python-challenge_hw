```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint
import seaborn as sns
import openweathermapy as owm


# Import API key
import api_keys
api_key = "c66cc81fe1c25a9497827e7957b2eee5"


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f"City count is " + str(len(cities)) + ".")   
```

```python
cities_df = pd.DataFrame(cities)
cities_df.head()
```

## Perform API Calls

```python
# OpenWeatherMap API Key
##api_key = api_keys.api_key
    
# Starting URL for Weather Map API Call  (I ADDED "&q")
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key + "&q="

#CREATE LISTS TO STORE THE DATA I AM LOOKING FOR 
latitude = []
longitude = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []

for x in range(len(cities)):
    cities_response = requests.get(f"{url}{cities[x]}").json()
    
    #SETTING CITIES IN API WITH NO INFO TO NAN 
    try:
        latitude.append(cities_response['coord']['lat'])
        longitude.append(cities_response['coord']['lon'])
        temperature.append(cities_response['main']['temp'])
        humidity.append(cities_response['main']['humidity'])
        wind_speed.append(cities_response['wind']['speed'])
        cloudiness.append(cities_response['clouds']['all'])
    except KeyError:
        temperature.append(np.nan)
        latitude.append(np.nan)
        longitude.append(np.nan)
        humidity.append(np.nan)
        wind_speed.append(np.nan)
        cloudiness.append(0)
           
    print(f"CITY: {x + 1} | {cities[x]}")
    print(f"{url}{cities[x]}")
        
```

```python
weather_info_df = pd.DataFrame({"City Name": cities,
                           "Temperature": temperature,
                           "Latitude": latitude,
                           "Longitude": longitude,
                           "Humidity (%)": humidity,
                           "Cloudiness (%)": cloudiness,
                           "Wind Speed (mph)": wind_speed,
                          })
weather_info_df = weather_info_df.dropna(how='any')
weather_info_df.head() 
```

```python
weather_info_df.to_csv("WeatherPy_Data.csv")
```

```python
#SCATTER PLOT - LATITUDE VS TEMP
plt.figure(figsize=(8,5))
plt.scatter(x = weather_info_df['Latitude'], y = weather_info_df['Temperature'], linewidths = 0.5,
           edgecolors = 'black', color = 'red', s=25)

plt.title(f"Latitude vs Temperature")
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.xlim(-60, 90)
plt.ylim(-10, 100)
sns.set()
plt.savefig('Latitude_v_Temperature.png')
plt.show()
```

```python
#SCATTER PLOT - LATITUDE VS HUMIDITY
plt.figure(figsize=(8,5))
plt.scatter(x = weather_info_df['Latitude'], y = weather_info_df['Humidity (%)'], linewidths = 0.5,
           edgecolors = 'black', color = 'red', s=25)

plt.title(f"Latitude vs Humidity")
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.xlim(-60, 90)
plt.ylim(0, 115)
sns.set()
plt.savefig('Latitude_v_Humidity.png')
plt.show()
```

```python
#SCATTER PLOT - LATITUDE VS CLOUDINESS
plt.figure(figsize=(8,5))
plt.scatter(x = weather_info_df['Latitude'], y = weather_info_df['Cloudiness (%)'], linewidths = 0.5,
           edgecolors = 'black', color = 'red', s=25)

plt.title(f"Latitude vs Cloudiness")
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.xlim(-60, 90)
plt.ylim(-10, 120)
sns.set()
plt.savefig("Latitude_v_Cloudiness.png")
plt.show()
```

```python
#SCATTER PLOT - LATITUDE VS WIND SPEED
plt.figure(figsize=(8,5))
plt.scatter(x = weather_info_df['Latitude'], y = weather_info_df['Wind Speed (mph)'], linewidths = 0.5,
           edgecolors = 'black', color = 'red', s=25)

plt.title(f"Latitude vs Wind Speed")
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.xlim(-60, 90)
plt.ylim(-5, 45)
sns.set()
plt.savefig("Latitude_v_WindSpeed.png")
plt.show()
```
